<h2><center> Criar Filtro para Detectar fronteiras na Imagem</center></h2>

```python
# Importa Bibliotecas
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np
%matplotlib inline
```

```python
# Ler imagem
image = mpimg.imread('images/dog.jpg')
plt.imshow(image)
```

### Converter a Imagem em grayscale - escala de cinza

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

### Criando o kernel

Um filtro muito comum para detectar bordas em imagens  o **Operador Sobel**.

* Pode ser usado para detectar padres de intensidade nas imagens
* Aplicar esse filtro  como se tivssimos tirando a derivada da imagem nas direes x e y.



<img src="images/sobel_ops.png" width=200 height=200, align="center">

```python
# 3x3 array para detectar bordas
# Mucanas VERTICAIS
S_x = np.array([[ -1, 0, 1], 
                   [ -2, 0, 2], 
                   [ -1, 0, 1]])

# Filtrando a imagem 
imagem_filtrada = cv2.filter2D(gray, -1, S_x)

plt.imshow(imagem_filtrada, cmap='gray')
```

```python
# Mucanas HORIZONTAIS
sobel_y = np.array([[ -1, -2, -1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])



# Filtrando imagem
filtered_image = cv2.filter2D(gray, -1, sobel_y)

plt.imshow(filtered_image, cmap='gray')
```

```python
# Mucanas HORIZONTAIS
sobel_y = np.array([[ 0, -1, 0], 
                   [ -1, 4, -1], 
                   [ 0, -1, 0]])



# Filtrando imagem
filtered_image = cv2.filter2D(gray, -1, sobel_y)

plt.imshow(filtered_image, cmap='gray')
```

```python
# Iluimnando a imagem
sobel_y = np.array([[ 0, -1, 0], 
                   [ -1, 8, -1], 
                   [ 0, -1, 0]])



# Filtrando imagem
filtered_image = cv2.filter2D(gray, -1, sobel_y)

plt.imshow(filtered_image, cmap='gray')
```

```python
# Escurecendo a imagem
sobel_y = np.array([[ 0, -1, 0], 
                   [ -1, 4, -2], 
                   [ 0, -1, 0]])



# Filtrando imagem
filtered_image = cv2.filter2D(gray, -1, sobel_y)

plt.imshow(filtered_image, cmap='gray')
```
